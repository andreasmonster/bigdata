```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
os.getcwd()
```

```python
#Reading .csv file containing our data
df = pd.read_csv('SBAnational.csv', sep=',', dtype='unicode')

#Displaying data table
df.head()
```

```python
value_counts = df['MIS_Status'].value_counts()


#plotting bar fo non-numeric values
ax = value_counts.plot(kind='bar', color='skyblue')


#Adding labels and titles
plt.xlabel('P I F = Paid in full & Charged Off')
plt.ylabel('Frequency')
plt.title('Target Variable MIS_Status')

for i, count in enumerate(value_counts):
    ax.text(i, count + 0.1, str(count), ha='center', va='bottom')
    
plt.show()
```

```python
# Assuming 'ApprovalFY' is a datetime column, if not, you might need to convert it
df['ApprovalFY'] = pd.to_datetime(df['ApprovalFY'], errors='coerce')

# Filter data for the desired range (1980s till 2014)
filtered_data = df[(df['ApprovalFY'].dt.year >= 1980) & (df['ApprovalFY'].dt.year <= 2014)]

# Count values in the filtered data
value_counts = filtered_data['ApprovalFY'].value_counts()

# Plotting bar for non-numeric values
ax = value_counts.plot(kind='line', color='skyblue')

# Adding labels and titles
plt.xlabel('Fiscal year of commitment')
plt.ylabel('Count of entities commitment')
plt.title('ApprovalFY')

for i, count in enumerate(value_counts):
    ax.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```
